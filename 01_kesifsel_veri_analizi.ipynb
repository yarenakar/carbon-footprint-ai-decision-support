```python
import zipfile
import warnings
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

warnings.filterwarnings("ignore")

DATA_PATH = "tekstil_karbon_datasetM.csv"

OUTPUT_DIR = Path("eda_outputs")
OUTPUT_DIR.mkdir(exist_ok=True)

plt.rcParams["figure.figsize"] = (9, 5)
plt.rcParams["axes.grid"] = True
plt.rcParams["font.size"] = 10


def save_fig(filename):
    path = OUTPUT_DIR / filename
    plt.tight_layout()
    plt.savefig(path, dpi=300, bbox_inches="tight")
    plt.show()
    print(f"Kaydedildi: {path}")


def check_columns(dataframe, columns):
    missing = [col for col in columns if col not in dataframe.columns]
    if missing:
        print(f"Uyarı: Şu kolonlar veri setinde bulunamadı: {missing}")
        return False
    return True


def slugify(text):
    replacements = {
        "ı": "i", "ş": "s", "ğ": "g", "ü": "u", "ö": "o", "ç": "c",
        "İ": "i", "Ş": "s", "Ğ": "g", "Ü": "u", "Ö": "o", "Ç": "c"
    }
    for k, v in replacements.items():
        text = text.replace(k, v)
    return text.lower().replace(" ", "_").replace("-", "").replace("/", "_")


def boxplot_by_category(dataframe, category_col, value_col, title, filename):
    if not check_columns(dataframe, [category_col, value_col]):
        return
    categories = dataframe[category_col].dropna().unique()
    data = [dataframe.loc[dataframe[category_col] == cat, value_col].dropna() for cat in categories]
    plt.figure()
    plt.boxplot(data, labels=categories, showmeans=True)
    plt.title(title)
    plt.xlabel(category_col)
    plt.ylabel(value_col)
    plt.xticks(rotation=30, ha="right")
    save_fig(filename)


def scatter_with_trend(dataframe, x_col, y_col, title, filename):
    if not check_columns(dataframe, [x_col, y_col]):
        return
    temp = dataframe[[x_col, y_col]].dropna()
    plt.figure()
    plt.scatter(temp[x_col], temp[y_col], alpha=0.6)
    if len(temp) > 1:
        z = np.polyfit(temp[x_col], temp[y_col], 1)
        p = np.poly1d(z)
        x_sorted = np.sort(temp[x_col])
        plt.plot(x_sorted, p(x_sorted), linewidth=2)
    corr = temp[x_col].corr(temp[y_col])
    plt.title(f"{title}\nKorelasyon: {corr:.3f}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    save_fig(filename)


def scatter_colored_by_category(dataframe, x_col, y_col, category_col, title, filename):
    if not check_columns(dataframe, [x_col, y_col, category_col]):
        return
    plt.figure()
    for cat in dataframe[category_col].dropna().unique():
        subset = dataframe[dataframe[category_col] == cat]
        plt.scatter(subset[x_col], subset[y_col], alpha=0.6, label=str(cat))
    plt.title(title)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.legend(title=category_col)
    save_fig(filename)


def save_table(table, filename):
    path = OUTPUT_DIR / filename
    table.to_csv(path, index=True, encoding="utf-8-sig")
    display(table)
    print(f"Tablo kaydedildi: {path}")


df = pd.read_csv(DATA_PATH, encoding="utf-8-sig")
print("Veri seti başarıyla okundu.")
print("Veri boyutu:", df.shape)

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()

treatment_cols = [
    "t_yenilenebilir_enerji",
    "t_yerel_tedarik",
    "t_dusuk_em_boya",
    "t_geri_donusum_prog",
    "t_makine_modernizasyon",
    "t_parti_optimizasyon"
]

existing_treatment_cols = [col for col in treatment_cols if col in df.columns]

general_info = pd.DataFrame({
    "Özellik": [
        "Satır sayısı",
        "Sütun sayısı",
        "Sayısal değişken sayısı",
        "Kategorik değişken sayısı",
        "Treatment değişken sayısı",
        "Tekrarlı satır sayısı",
        "batch_id benzersiz mi?"
    ],
    "Değer": [
        df.shape[0],
        df.shape[1],
        len(numeric_cols),
        len(categorical_cols),
        len(existing_treatment_cols),
        df.duplicated().sum(),
        df["batch_id"].is_unique if "batch_id" in df.columns else "batch_id yok"
    ]
})

save_table(general_info, "01_veri_genel_yapi.csv")

variable_groups = {
    "Kimlik Değişkeni": ["batch_id"],
    "Ürün ve Üretim Özellikleri": [
        "urun_tipi", "kumas_tipi", "parti_buyuklugu",
        "kumas_gramaj_g_m2", "kimyasal_kg_adet",
        "uretim_suresi_dk_adet", "su_tuketim_lt_adet"
    ],
    "Operasyonel Değişkenler": [
        "ay", "vardiya_sayisi", "kapasite_kullanim_pct", "makine_yasi_yil"
    ],
    "Scope 1 - Yakıt Girdileri": [
        "dogalgaz_m3", "fuel_oil_lt", "lp_gaz_kg"
    ],
    "Scope 2 - Elektrik Girdileri": [
        "elektrik_kwh", "yenilenebilir_oran_pct", "ef_elektrik"
    ],
    "Scope 3 - Tedarik, Atık ve Atık Su Girdileri": [
        "hammadde_mesafe_km", "tasima_modu", "hammadde_ton",
        "tekstil_atik_kg", "geri_donusum_orani_pct", "atiksu_lt"
    ],
    "Tedavi / Müdahale Değişkenleri": treatment_cols,
    "Hesaplanmış Emisyon Değişkenleri": [
        "scope1_kgCO2e", "scope2_kgCO2e",
        "scope3_tasima_kgCO2e", "scope3_atik_kgCO2e",
        "scope3_atiksu_kgCO2e", "scope3_kimyasal_kgCO2e",
        "scope3_kgCO2e", "toplam_karbon_kgCO2e"
    ]
}

rows = []
for group, columns in variable_groups.items():
    for col in columns:
        rows.append({
            "Değişken Grubu": group,
            "Değişken": col,
            "Veri Setinde Var mı?": col in df.columns
        })

variable_group_table = pd.DataFrame(rows)
save_table(variable_group_table, "02_degisken_gruplari.csv")

descriptive_stats = df[numeric_cols].describe().T
descriptive_stats["median"] = df[numeric_cols].median()
descriptive_stats = descriptive_stats[
    ["count", "mean", "median", "std", "min", "25%", "50%", "75%", "max"]
]

save_table(descriptive_stats.round(3), "05_sayisal_degisken_tanimlayici_istatistikler.csv")

target_col = "toplam_karbon_kgCO2e"

if check_columns(df, [target_col]):
    plt.figure()
    plt.hist(df[target_col].dropna(), bins=30)
    plt.title("Toplam Karbon Emisyonu Dağılımı")
    plt.xlabel("Toplam Karbon Emisyonu (kgCO2e)")
    plt.ylabel("Frekans")
    save_fig("06_toplam_karbon_histogram.png")

    plt.figure()
    plt.boxplot(df[target_col].dropna(), orientation="vertical", showmeans=True)
    plt.title("Toplam Karbon Emisyonu Boxplot")
    plt.ylabel("Toplam Karbon Emisyonu (kgCO2e)")
    save_fig("06_toplam_karbon_boxplot.png")

scope_cols = ["scope1_kgCO2e", "scope2_kgCO2e", "scope3_kgCO2e"]

if check_columns(df, scope_cols + ["toplam_karbon_kgCO2e"]):
    scope_sums = df[scope_cols].sum()
    scope_shares = (scope_sums / df["toplam_karbon_kgCO2e"].sum() * 100).round(2)

    plt.figure()
    scope_shares.plot(kind="bar")
    plt.title("Toplam Karbon Emisyonunun Scope Bazlı Payı")
    plt.xlabel("Scope")
    plt.ylabel("Pay (%)")
    plt.xticks(rotation=0)
    save_fig("07_scope_paylari_bar_chart.png")

    plt.figure()
    scope_shares.plot(kind="pie", autopct="%1.1f%%", startangle=90)
    plt.title("Scope Bazlı Emisyon Payları")
    plt.ylabel("")
    save_fig("07_scope_paylari_pie.png")

boxplot_by_category(
    df,
    "urun_tipi",
    "toplam_karbon_kgCO2e",
    "Ürün Tipine Göre Toplam Karbon Dağılımı",
    "08_urun_tipine_gore_karbon_boxplot.png"
)

boxplot_by_category(
    df,
    "kumas_tipi",
    "toplam_karbon_kgCO2e",
    "Kumaş Tipine Göre Toplam Karbon Dağılımı",
    "09_kumas_tipine_gore_karbon_boxplot.png"
)

scatter_with_trend(
    df,
    "parti_buyuklugu",
    "toplam_karbon_kgCO2e",
    "Parti Büyüklüğü - Toplam Karbon İlişkisi",
    "10_parti_buyuklugu_toplam_karbon_scatter.png"
)

scatter_with_trend(
    df,
    "elektrik_kwh",
    "scope2_kgCO2e",
    "Elektrik Tüketimi - Scope 2 Emisyonu İlişkisi",
    "11_elektrik_scope2_scatter_destek.png"
)

scatter_colored_by_category(
    df,
    "hammadde_mesafe_km",
    "scope3_tasima_kgCO2e",
    "tasima_modu",
    "Hammadde Mesafesi - Scope 3 Taşıma Emisyonu İlişkisi (Taşıma Moduna Göre Renkli)",
    "12_hammadde_mesafe_scope3_tasima_scatter_destek.png"
)

scatter_with_trend(
    df,
    "geri_donusum_orani_pct",
    "scope3_atik_kgCO2e",
    "Geri Dönüşüm Oranı - Atık Kaynaklı Emisyon İlişkisi",
    "13_geri_donusum_scope3_atik_scatter.png"
)

scatter_matrix_cols = [
    "kimyasal_kg_adet",
    "su_tuketim_lt_adet",
    "atiksu_lt",
    "scope3_kimyasal_kgCO2e",
    "scope3_atiksu_kgCO2e"
]

if check_columns(df, scatter_matrix_cols):
    scatter_matrix(
        df[scatter_matrix_cols].dropna(),
        figsize=(12, 12),
        diagonal="hist",
        alpha=0.6
    )
    plt.suptitle("Kimyasal, Su ve Atık Su Değişkenleri Scatter Plot Matrisi", y=1.02)
    save_fig("14_kimyasal_su_atiksu_scatter_matrix.png")

scatter_with_trend(
    df,
    "makine_yasi_yil",
    "elektrik_kwh",
    "Makine Yaşı - Elektrik Tüketimi İlişkisi",
    "15_makine_yasi_elektrik_scatter.png"
)

boxplot_by_category(
    df,
    "t_makine_modernizasyon",
    "elektrik_kwh",
    "Makine Modernizasyonuna Göre Elektrik Tüketimi",
    "15_modernizasyon_elektrik_boxplot.png"
)

treatment_scope_pairs = [
    ("t_geri_donusum_prog", "scope3_atik_kgCO2e", "Geri Donusum Programi - Scope 3 Atik"),
    ("t_parti_optimizasyon", "scope3_atik_kgCO2e", "Parti Optimizasyonu - Scope 3 Atik"),
    ("t_yerel_tedarik", "scope3_tasima_kgCO2e", "Yerel Tedarik - Scope 3 Tasima"),
    ("t_yenilenebilir_enerji", "scope2_kgCO2e", "Yenilenebilir Enerji - Scope 2")
]

for treatment_col, outcome_col, title in treatment_scope_pairs:
    if check_columns(df, [treatment_col, outcome_col]):
        safe_title = slugify(title)
        boxplot_by_category(
            df,
            treatment_col,
            outcome_col,
            title,
            f"17_{safe_title}_boxplot.png"
        )

corr_cols = [
    "parti_buyuklugu",
    "kumas_gramaj_g_m2",
    "kimyasal_kg_adet",
    "uretim_suresi_dk_adet",
    "su_tuketim_lt_adet",
    "kapasite_kullanim_pct",
    "makine_yasi_yil",
    "dogalgaz_m3",
    "elektrik_kwh",
    "hammadde_mesafe_km",
    "hammadde_ton",
    "tekstil_atik_kg",
    "geri_donusum_orani_pct",
    "atiksu_lt",
    "toplam_karbon_kgCO2e"
]

existing_corr_cols = [col for col in corr_cols if col in df.columns]

if len(existing_corr_cols) > 1:
    corr_matrix = df[existing_corr_cols].corr()

    plt.figure(figsize=(13, 10))
    im = plt.imshow(corr_matrix, aspect="auto", vmin=-1, vmax=1, cmap="coolwarm")
    plt.colorbar(im, fraction=0.046, pad=0.04)
    plt.title("Seçilmiş Sayısal Değişkenler Korelasyon Matrisi")
    plt.xticks(range(len(existing_corr_cols)), existing_corr_cols, rotation=90)
    plt.yticks(range(len(existing_corr_cols)), existing_corr_cols)

    for i in range(len(existing_corr_cols)):
        for j in range(len(existing_corr_cols)):
            plt.text(j, i, f"{corr_matrix.iloc[i, j]:.2f}", ha="center", va="center", fontsize=7)

    save_fig("18_korelasyon_heatmap.png")

outlier_cols = [
    "toplam_karbon_kgCO2e",
    "parti_buyuklugu",
    "elektrik_kwh",
    "hammadde_mesafe_km",
    "tekstil_atik_kg",
    "atiksu_lt"
]

existing_outlier_cols = [col for col in outlier_cols if col in df.columns]

outlier_rows = []

for col in existing_outlier_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()

    outlier_rows.append({
        "Değişken": col,
        "Q1": Q1,
        "Q3": Q3,
        "IQR": IQR,
        "Alt Sınır": lower_bound,
        "Üst Sınır": upper_bound,
        "Aykırı Değer Sayısı": outlier_count,
        "Aykırı Değer Oranı (%)": round(outlier_count / len(df) * 100, 2)
    })

    plt.figure()
    plt.boxplot(df[col].dropna(), orientation="vertical", showmeans=True)
    plt.title(f"Aykırı Değer Kontrolü - {col}")
    plt.ylabel(col)
    save_fig(f"19_aykiri_deger_{col}.png")

outlier_table = pd.DataFrame(outlier_rows)
save_table(outlier_table.round(3), "19_aykiri_deger_iqr_tablosu.csv")

zip_filename = "eda_outputs.zip"

with zipfile.ZipFile(zip_filename, "w", zipfile.ZIP_DEFLATED) as zipf:
    for file in OUTPUT_DIR.iterdir():
        zipf.write(file, arcname=file.name)

print(f"Tüm EDA çıktıları zip dosyasına kaydedildi: {zip_filename}")
```
